```python
# File1: Canada

File1 = pd.read_csv('CAvideos.csv')

# Add a new column named "region": indicate which region the data comes from

File1['region'] = 'CA'

# Add a new column named "category": map the category_id to the categories

File1['category_id'] = File1['category_id'].astype(str)
categories_File1 = pd.read_json('CA_category_id.json')

id_to_category = {}
for category in categories_File1['items']:
    id_to_category[category['id']] = category['snippet']['title']

File1.insert(5, 'category', File1['category_id'].map(id_to_category))
File1.to_csv('Data Pre-processing/CAvideos.csv')
```

```python
# Delete unnecessary columns

data = File.drop(['Unnamed: 0', 'thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed'], axis=1)
```

```python
data.head()
```

```python
print(data.shape)
print(data.nunique())
```

```python
# Preview default values and data types

data.info()
```

```python
# Fix date format

data['trending_date'] = pd.to_datetime(data['trending_date'], format = '%y.%d.%m')
data['publish_time'] = pd.to_datetime(data['publish_time'], format = '%Y-%m-%dT%H:%M:%S.%fZ')
data[['trending_date','publish_time']].head()
```

```python
# Converts the values of columns 'views', 'likes', 'dislikes', and 'comment_count' from float to int

type_int_list = ['views', 'likes', 'dislikes', 'comment_count']
for column in type_int_list:
    data[column] = data[column].astype(int)

# Converts the values of columns 'category_id' from int to string
    
type_str_list = ['category_id']
for column in type_str_list:
    data[column] = data[column].astype(str)
```
